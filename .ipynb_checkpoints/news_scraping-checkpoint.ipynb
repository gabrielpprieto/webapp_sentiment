## News API

```python
# instantiate newsapi object
newsapi = NewsApiClient(api_key='f418175c97314f91ad85efb39a9b1948')

# instantiate dataframe
df = pd.DataFrame()

# instantitate list with tickers
search_terms = ['TSLA', 'Tesla', 'Elon Musk']

# loop through all tickers
for search_term in search_terms:
    
    # get articles from first 5 pages - News API limit for Free Tier
    for page in tqdm(range(1, 6)):
        
        # request information
        all_articles = newsapi.get_everything(q=search_term, # search_term
                                              from_param=str(datetime.today()-timedelta(1))[:10], # yesterday
                                              to=str(datetime.today())[:10], # today
                                              language='en', # english
                                              page=page) # for 5 pages - Limit

        # create dataframe with information
        i_df = pd.DataFrame(all_articles['articles'])
        
        # add ticker column
        i_df['search_term'] = search_term

        # concatenate dataframes for all pages
        df = pd.concat([df, i_df], axis=0)

# reset dataframe's index
df = df.reset_index(drop=True)

# format source column to string
df['source'] = df['source'].map(lambda x: x['name'])
```

```python
# Connetction to DataBase
conn = psycopg2.connect(host='192.168.0.103',
                        user='gpp',
                        password='gppfp2123',
                        dbname='news')

# Set User Agent
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) \
AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.61 Safari/537.36'}
     
# Loop Though all Pages available
for page in tqdm(range(1,2001)):

    # Set Wikipedia url to scrape the table from
    url = 'https://seekingalpha.com/market-news/' + str(page)

    # Set User Agent
#     headers = {'User-Agent': user_agents[page][0]}
    
    # Request from url
    res = requests.get(url, headers=headers)

    # Create soup object from the home page
    soup = BeautifulSoup(res.content, 'lxml')

    # Find all anchor tags with specific parameters
    table = soup.find(name="ul", attrs={'class':"item-list",'id':'latest-news-list'})
    
    # Try to Get Content
    try:
        # Loop through table rows  
        for li in table.find_all('li'):
                        
            # Link for each news page
            link = 'https://seekingalpha.com' + li.find('a').attrs['href']

            # Request from url
            res = requests.get(link, headers=headers)

            # Create soup object from the home page
            soup = BeautifulSoup(res.content, 'lxml')

            # Find all anchor tags with specific parameters
            bullets = soup.find(name='div', attrs={'id': 'bullets_ul'})

            text = ''
            for p in bullets.find_all('p'):
                if len(bullets.find_all('p')) > 1:
                    text = text + p.text
                else:
                    text = p.text
        
            with conn.cursor() as cur:

                # SQL query
                query = f"INSERT INTO seeking_alpha (id, date, header, link, content) VALUES ($${li['id']}$$, $${li['data-last-date']}$$, $${li.find('a').text}$$, $${str('https://seekingalpha.com') + li.find('a').attrs['href']}$$, $${text}$$)"

                # Execute Query
                cur.execute(query)

            # Commit to Modify the DataBase
            conn.commit()
           
            # Sleep
            time.sleep(random.randint(5,15) + np.exp(random.random()))
            time.sleep(random.randint(2,5) + np.exp(random.random()))

    # Exception Pass 
    except AttributeError:
        time.sleep(random.randint(30,45) + np.exp(random.random()))
        continue
```
